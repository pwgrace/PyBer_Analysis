```python
%matplotlib inline
```

```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston',marker="*", color="blue", linewidth=2)
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
fig, ax=plt.subplots()
ax.plot(x_axis, y_axis, label='Boston',color='green',marker='D',linewidth=2)
ax.legend()
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Data")
ax.set_ylabel("Fare ($)")
ax.set_ylim(0,45)
ax.grid()
```

```python
# Create the plot
plt.bar(x_axis, y_axis)
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot
plt.barh(x_axis, y_axis, color="magenta", label="Boston")
plt.gca().invert_yaxis()
plt.xlabel("Fare ($)")
plt.ylabel("Date")
plt.title("PyBer Fare by Month")
plt.legend()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color='cyan', label='Chicago')
ax.set_xlabel('Fare ($)')
ax.set_ylabel('Date')
ax.set_title('PyBer Fare by Month')
ax.legend()
plt.gca().invert_yaxis()
```

```python
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(y_axis, x_axis,color='red',label='Chicago')
plt.legend()
plt.title("PyBer Fare by Month")
plt.xlabel("Fare ($)")
plt.ylabel("Date")
plt.xlim(0,50)
plt.gca().invert_yaxis()
```

```python
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)

plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=[i*5 for i in y_axis], color='skyblue', edgecolor='black', linewidths=2, alpha=0.2, label='Boston')
ax.set_title('PyBer Fare by Month')
ax.set_xlabel('Fare ($)')
ax.set_ylabel('Date')
ax.set_xlim(0,50)
ax.legend()
plt.gca().invert_yaxis()
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
        explode=explode_values, 
        colors=colors, 
        labels=x_axis, 
        autopct='%.1f%%')
plt.show()
```

```python
explode=[0,0,0.2,0,0,0,0.2,0,0,0,0,0]
fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(y_axis,
       labels=x_axis,
       explode=explode,
       colors=colors,
       shadow=True,
       startangle=265,
      autopct='%.1f%%')
plt.gca().invert_yaxis()
plt.show()
```
